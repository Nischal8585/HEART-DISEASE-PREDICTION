```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd

# Assuming `heart_data` is your DataFrame
# Splitting features and target
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']

# Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

# Initialize the Random Forest Classifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=2)

# Train the model
random_forest.fit(X_train, Y_train)

# Evaluate the model on training and test sets
Y_train_pred = random_forest.predict(X_train)
Y_test_pred = random_forest.predict(X_test)



# Feature Importance
feature_importance = random_forest.feature_importances_

# Create a DataFrame to display feature importance
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
})

# Sort by importance
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Display feature importance
print("\nFeature Importance:\n", importance_df)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.title('Feature Importance (Random Forest)')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.gca().invert_yaxis()  # Most important feature on top
plt.show()
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.svm import SVC  # Import the SVM model class

model = SVC(kernel='linear', probability=True, random_state=42)
```

```python

model.fit(X_train, Y_train)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy of the model : ', test_data_accuracy)
```
